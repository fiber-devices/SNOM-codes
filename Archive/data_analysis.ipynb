# Standing wave pitch analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Forward and backward analysis

```python
# Load data 

# forward
general_path_f = './data/20250815_201926_forward/'
general_file_name = '20250815_0um'
extension = '.csv'

file_path_f = general_path_f + general_file_name + extension 
data_f = pd.read_csv(file_path_f)

# backward
general_path_b = './data/20250815_202008_backward/'
file_path_b = general_path_b + general_file_name + extension 
data_b = pd.read_csv(file_path_b)
```

```python
x_f = data_f.iloc[:,0]
V_f = data_f.iloc[:,-1]

x_b = data_b.iloc[:,0]
V_b = data_b.iloc[:,-1]

plt.figure(figsize=(10, 5))

plt.plot(x_f, V_f, '.-', label='Forward')


plt.plot(x_b, V_b, '.-', label='Backward')
plt.xlabel('x position (um)')
plt.ylabel('Voltage (V)')
plt.legend()
plt.grid(True)
plt.show()

```

#### Repeated scans #1

```python
# Load data
num_len = 10

# setting path
general_path = './data/20250815_201105_backwardrepeat/'
general_file_name = '20250815_'
extension = '.csv'

class Scan_Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

Scan_num = []

for i in range(num_len):
    file_path = general_path + general_file_name + str(i) + 'um' + extension 
    data = pd.read_csv(file_path)
    x_array = data.iloc[:,0]
    V_array = data.iloc[:,-1]

    Scan_num.append(Scan_Point(x_array, V_array))

```

```python
# Plot data 

plt.figure(figsize=(10, 5))
for i in range(num_len):
    plt.plot(Scan_num[i].x, Scan_num[i].y, '.-')

plt.xlabel('x position (um)')
plt.ylabel('Voltage (V)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np

# --- Load CSV and skip unit row (row 1) ---
csv_file = "20251007_kopzt.csv"
data_kopzt = pd.read_csv(
    csv_file,
    skiprows=[1],  #  skip the unit row like 's','V','V'
    dtype={"Time": np.float64, "Channel C": np.float64, "Channel D": np.float64}
)


# --- Extract numeric arrays ---
time_data = data_kopzt["Time"].to_numpy()
channel_c_data = data_kopzt["Channel C"].to_numpy()
channel_d_data = data_kopzt["Channel D"].to_numpy()

```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time_data, channel_c_data, label="PZT driver")
plt.plot(time_data, channel_d_data, label="Interferometer")

# Labeling the axes
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.xlim(0, 0.2)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Time (s)", fontsize=14)
plt.ylabel("Voltage (V)", fontsize=14)



# Adding a legend
plt.legend()

# Grid and show
plt.grid(True)
plt.show()
```
